```python
# Import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
%matplotlib inline
```

```python
# Read the data
data = pd.read_excel('data_v1.xlsx')
data
```

```python
# Set the time index
data = data.set_index(data['time'])
data
```

```python
pl = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
```

```python
# Plot the time seris Beijing
fig = plt.figure(figsize=(16, 10), dpi=120)
grid = plt.GridSpec(2, 3)

#plt.title('Time series of concentrations of air pollutions in Beijing')
plt.rcParams['font.sans-serif']=['Times New Roman']

# PM2.5
plt.subplot(grid[0, 0])
plt.plot(data['PM2.5_BJ'], color='r')
plt.title(pl[0])
plt.xlabel('Time')
plt.ylabel('concentration')

plt.xticks(fontsize=12, rotation=0, ha='center', va='top')
plt.yticks(fontsize=12, rotation=0, ha='right', va='center')
plt.tick_params(axis='both', bottom=True, top=False, left=True, right=False, 
                direction='in', which='major')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# PM10
plt.subplot(grid[0, 1])
plt.plot(data['PM10_BJ'], color='r')
plt.title(pl[1])
plt.xlabel('Time')
plt.ylabel('concentration')

plt.xticks(fontsize=12, rotation=0, ha='center', va='top')
plt.yticks(fontsize=12, rotation=0, ha='right', va='center')
plt.tick_params(axis='both', bottom=True, top=False, left=True, right=False, 
                direction='in', which='major')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# SO2
plt.subplot(grid[0, 2])
plt.plot(data['SO2_BJ'], color='r')
plt.title(pl[2])
plt.xlabel('Time')
plt.ylabel('concentration')

plt.xticks(fontsize=12, rotation=0, ha='center', va='top')
plt.yticks(fontsize=12, rotation=0, ha='right', va='center')
plt.tick_params(axis='both', bottom=True, top=False, left=True, right=False, 
                direction='in', which='major')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# NO2
plt.subplot(grid[1, 0])
plt.plot(data['NO2_BJ'], color='r')
plt.title(pl[3])
plt.xlabel('Time')
plt.ylabel('concentration')

plt.xticks(fontsize=12, rotation=0, ha='center', va='top')
plt.yticks(fontsize=12, rotation=0, ha='right', va='center')
plt.tick_params(axis='both', bottom=True, top=False, left=True, right=False, 
                direction='in', which='major')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# CO
plt.subplot(grid[1, 1])
plt.plot(data['CO-95per_BJ'], color='r')
plt.title(pl[4])
plt.xlabel('Time')
plt.ylabel('concentration')

plt.xticks(fontsize=12, rotation=0, ha='center', va='top')
plt.yticks(fontsize=12, rotation=0, ha='right', va='center')
plt.tick_params(axis='both', bottom=True, top=False, left=True, right=False, 
                direction='in', which='major')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# O3
plt.subplot(grid[1, 2])
plt.plot(data['O3-8H-90per_BJ'], color='r')
plt.title(pl[5])
plt.xlabel('Time')
plt.ylabel('concentration')

plt.xticks(fontsize=12, rotation=0, ha='center', va='top')
plt.yticks(fontsize=12, rotation=0, ha='right', va='center')
plt.tick_params(axis='both', bottom=True, top=False, left=True, right=False, 
                direction='in', which='major')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

```python
# Calculate IAQI
a = [[0, 35, 75, 115, 150, 250, 350, 500],
     [0, 50, 150, 250, 350, 420, 500, 600],
     [0, 50, 150, 475, 800, 1600, 2100, 2620],
     [0, 40, 80, 180, 280, 565, 750, 940],
     [0, 2, 4, 14, 24, 36, 48, 60],
     [0, 100, 160, 215, 265, 800, 1000, 1200]]
b = [0, 50, 100, 150, 200, 300, 400, 500]


col = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO-95per', 'O3-8H-90per']
link = '_'
cities = ['BJ', 'SH', 'GZ', 'SZ']

s1 = 'IAQI'
for i in range(4):
    for j in range(6):
        s2 = col[j]+link+cities[i]
        s3 = s1+pl[j]+cities[i]
        data[s3] = [0]*60
        for k in range(7):
            data[s3].loc[(data[s2] >= a[j][k]) & (data[s2] < a[j][k+1])] = (
                (data[s2] - a[j][k])*(b[k+1] - b[k])/(a[j][k+1] - a[j][k]) + b[k])
        
data
```

```python
# Dominant pollution
for i in range(4):
    s4 = 'AQI'+cities[i]
    s5 = 'pollution'+cities[i]
    data[s4] = [0]*60
    data[s5] = ''
    for j in range(6):
        s3 = s1+pl[j]+cities[i]
        data[s5].loc[data[s3] > data[s4]] = pl[j]
        data[s4].loc[data[s3] > data[s4]] = data[s3]
        
data[['pollutionBJ', 'pollutionSH', 'pollutionGZ', 'pollutionSZ']]        
```

```python
# Level
l = [0, 50, 100, 150, 200, 300]
level = ['Excellent', 'Good', 'Mild', 'Moderate', 'Severe']

for i in range(4):
    s4 = 'AQI'+cities[i]
    s6 = 'level'+cities[i]
    data[s6] = 'Hazard'*60
    for j in range(5):
        data[s6].loc[(data[s4] > l[j]) & (data[s4] <= l[j+1])] = level[j]
        
data[['levelBJ', 'levelSH', 'levelGZ', 'levelSZ']]
```

```python
def survey(df):
    labels = df.index.tolist() #æå–åˆ†ç±»æ˜¾ç¤ºæ ‡ç­¾
    results = df.to_dict(orient = 'list') #å°†æ•°å€¼ç»“æžœè½¬åŒ–ä¸ºå­—å…¸
    category_names = list(results.keys()) # æå–å­—å…¸é‡Œé¢çš„ç±»åˆ«ï¼ˆé”®-keyï¼‰
    data = np.array(list(results.values())) #æå–å­—å…¸é‡Œé¢çš„æ•°å€¼ï¼ˆå€¼-valueï¼‰

    category_colors = plt.get_cmap('RdYlGn')(np.linspace(0.15, 0.85, data.shape[0])) 
    #è®¾ç½®å æ¯”æ˜¾ç¤ºçš„é¢œè‰²ï¼Œå¯ä»¥è‡ªå®šä¹‰ï¼Œä¿®æ”¹æ‹¬å·é‡Œé¢çš„å‚æ•°å³å¯ï¼Œå¦‚ä¸‹
	#category_colors = plt.get_cmap('hot')(np.linspace(0.15, 0.85, data.shape[0]))
	
	
    fig, ax = plt.subplots(figsize=(12, 9)) #åˆ›å»ºç”»å¸ƒï¼Œå¼€å§‹ç»˜å›¾
    ax.invert_xaxis()#è¿™ä¸ªå¯ä»¥é€šè¿‡è®¾ç½®dfä¸­columnsçš„é¡ºåºè°ƒæ•´
    ax.yaxis.set_visible(False) #è®¾ç½®yè½´åˆ»åº¦ä¸å¯è§
    ax.set_xticklabels(labels=labels, rotation=90) #æ˜¾ç¤ºxè½´æ ‡ç­¾ï¼Œå¹¶æ—‹è½¬90åº¦
    ax.set_ylim(0,1) #è®¾ç½®yè½´çš„æ˜¾ç¤ºèŒƒå›´
    starts = 0  #ç»˜åˆ¶åŸºå‡†
    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        heights = data[i,: ]/ data.sum(axis =0) #è®¡ç®—å‡ºæ¯æ¬¡éåŽ†æ—¶å€™çš„ç™¾åˆ†æ¯”
        ax.bar(labels, heights, bottom=starts, width=0.5,label=colname, color=color,edgecolor ='gray') # ç»˜åˆ¶æŸ±çŠ¶å›¾
        xcenters = starts + heights/2  #è¿›è¡Œæ–‡æœ¬æ ‡è®°ä½ç½®çš„é€‰å®š
        starts += heights  #æ ¸å¿ƒä¸€æ­¥ï¼Œå°±æ˜¯åŸºäºŽåŸºå‡†ä¸Šçš„ç™¾åˆ†æ¯”ç´¯åŠ 
        #print(starts) è¿™ä¸ªå˜é‡å°±æ˜¯èƒ½å¦ç™¾åˆ†æ¯”æ˜¾ç¤ºçš„å…³é”®ï¼Œå¯ä»¥æ‰“å°è¾“å‡ºçœ‹ä¸€ä¸‹
        percentage_text = data[i,: ]/ data.sum(axis =0) #æ–‡æœ¬æ ‡è®°çš„æ•°æ®
        
        r, g, b, _ = color  # è¿™é‡Œè¿›è¡Œåƒç´ çš„åˆ†å‰²
        text_color = 'white' if r * g * b < 0.5 else 'k'  #æ ¹æ®é¢œè‰²åŸºè°ƒåˆ†é…æ–‡æœ¬æ ‡è®°çš„é¢œè‰²
        for y, (x, c) in enumerate(zip(xcenters, percentage_text)):
            ax.text(y, x, f'{round(c*100,2)}%', ha='center', va='center',
                    color=text_color, rotation = 90) #æ·»åŠ æ–‡æœ¬æ ‡è®°
    ax.legend(ncol=len(category_names), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='large') #è®¾ç½®å›¾ä¾‹
    return fig, ax  #è¿”å›žå›¾åƒ
```

```python
# Plot contributionï¼ˆå¾…ä¿®æ”¹ï¼‰ï¼ˆæœªçŸ¥æ˜¯å¦ç”¨ï¼Œè¿è¡Œå¤±è´¥ï¼‰
# encouraged at https://matplotlib.org/stable/gallery/lines_bars_and_markers/horizontal_barchart_distribution.html#sphx-glr-gallery-lines-bars-and-markers-horizontal-barchart-distribution-py
# and https://blog.csdn.net/lys_828/article/details/106524459

def survey(df):

    labels = df.index.tolist()
    results = df.to_dict(orient = 'list')
    category_names = list(results.keys())
    data = np.array(list(results.values()))

    category_colors = plt.get_cmap('RdYlGn')(
        np.linspace(0.15, 0.85, data.shape[0]))

    fig, ax = plt.subplots(figsize=(8, 5))
    ax.invert_xaxis()
    ax.xaxis.set_visible(False)
    ax.set_xticklabels(labels=labels)
    ax.set_ylim(0, 1)
    starts = 0

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        heights = data[i,:]/data.sum(axis=0)
        ax.bar(labels, 0.5, heights, bottom=starts,
               label=colname, color=color)
        xcenters = starts + heights / 2
        starts += heights
        percentage_text = data[i,:]/data.sum(axis=0)

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        for y, (x, c) in enumerate(zip(xcenters, percentage_text)):
            ax.text(y, x, f'{round(c*100,2)}%', ha='center', va='center',
                    color=text_color)
    ax.legend(ncol=len(category_names), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='small')

    return fig, ax
```

```python
category_names = ['Strongly disagree', 'Disagree',
                  'Neither agree nor disagree', 'Agree', 'Strongly agree']
results = {
    'Question 1': [10, 15, 17, 32, 26],
    'Question 2': [26, 22, 29, 10, 13],
    'Question 3': [35, 37, 7, 2, 19],
    'Question 4': [32, 11, 9, 15, 33],
    'Question 5': [21, 29, 5, 5, 40],
    'Question 6': [8, 19, 5, 30, 38]
}
df = pd.DataFrame(results,index = category_names)

survey(df)
plt.show()
```

```python
# Plot 3-dï¼ˆå¾…ä¿®æ”¹ï¼‰ï¼ˆxæ—¶é—´yåŸŽå¸‚zç©ºæ°”è´¨é‡ï¼‰
# Fixing random state for reproducibility
np.random.seed(19680801)


fig = plt.figure()
ax = fig.add_subplot(projection='3d')

colors = ['r', 'g', 'b', 'y']
#æ”¹ä¸º4åŸŽå¸‚åç§°
yticks = [3, 2, 1, 0]
for c, k in zip(colors, yticks):
    # Generate the random data for the y=k 'layer'.
    xs = np.arange(20)
    ys = np.random.rand(20)

    # You can provide either a single color or an array with the same length as
    # xs and ys. To demonstrate this, we color the first bar of each set cyan.
    cs = [c] * len(xs)
    cs[0] = 'c'

    # Plot the bar graph given by xs and ys on the plane y=k with 80% opacity.
    ax.bar(xs, ys, zs=k, zdir='y', color=cs, alpha=0.8)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# On the y axis let's only label the discrete values that we have data for.
ax.set_yticks(yticks)

plt.show()
```
