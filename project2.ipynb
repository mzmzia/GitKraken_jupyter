```python
# Import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the data
df = pd.read_excel('data_v2.xlsx')
df
```

```python
df = df.set_index(df['time'])
df
```

```python
pl = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
col = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO-95per', 'O3-8H-90per']
link = '_'
scities = ['BJ', 'SH', 'GZ', 'SZ']
cities = ['Beijing', 'Shanghai', 'Guangzhou', 'Shenzhen']
colors = ['r', 'y', 'b', 'g']
# CO and O3 have no annual average limit, but daily limit
level1 = [15, 40, 20, 40, 4, 100]
level2 = [35, 70, 60, 40, 4, 160]
```

```python
df
```

```python
Adf = pd.DataFrame(df.resample('Y').mean()['IAQIPM2.5BJ'])
```

```python
for i in range(4):
    s1 = 'AQI'+scities[i]
    Adf[s1] = pd.DataFrame(df.resample('Y').mean()[s1])
    for j in range(6):
        s2 = 'IAQI'+pl[j]+scities[i]
        Adf[s2] = pd.DataFrame(df.resample('Y').mean()[s2])
        
Adf
```

```python
Adf.to_excel('data_annual.xlsx')
```

```python
Mdf = pd.DataFrame(df.groupby(['month']).mean()['IAQIPM2.5BJ'])
Mdf
```

```python
for i in range(4):
    s1 = 'AQI'+scities[i]
    Mdf[s1] = pd.DataFrame(df.groupby(['month']).mean()[s1])
    for j in range(6):
        s2 = 'IAQI'+pl[j]+scities[i]
        Mdf[s2] = pd.DataFrame(df.groupby(['month']).mean()[s2])
        
Mdf
```

```python
Mdf.to_excel('data_monthly.xlsx')
```

```python
# Plot the annual average

fig = plt.figure(figsize=(25, 12), dpi=120)
grid = plt.GridSpec(2, 3)

plt.rcParams['font.sans-serif']=['Times New Roman']

for i in range(6):
    plt.subplot(grid[int(i/3), i%3])
    plt.title('Annual average concentration of ' + pl[i])
    plt.xlabel('Year')
    plt.ylabel('Concentration')
    for j in range(4):
        plt.plot(df.resample('Y').mean()[col[i]+link+scities[j]], 
                 color=colors[j], label=cities[j])
    
    plt.axhline(y=level1[i], ls='-', c='darkviolet', alpha=0.7, label='Level-1 annual limit')
    plt.axhline(y=level2[i], ls='--', c='darkviolet', alpha=0.7, label='Level-2 annual limit') 
    
    
plt.legend()
    
plt.show()
```

```python
# Plot climatology

fig = plt.figure(figsize=(25, 12), dpi=120)
grid = plt.GridSpec(2, 3)

plt.rcParams['font.sans-serif']=['Times New Roman']

for i in range(6):
    plt.subplot(grid[int(i/3), i%3])
    plt.title('Climatology of ' + pl[i])
    plt.xlabel('Year')
    plt.ylabel('Concentration')
    for j in range(4):
        plt.plot(df.groupby(['month']).mean()[col[i]+link+scities[j]], 
                 color=colors[j], label=cities[j]) 
    
plt.legend()
    
plt.show()
```

```python
from itertools import compress
#scities = ['BJ', 'SH', 'GZ', 'SZ']
colors = ['green', 'yellow', 'orange', 'red', 'purple']
sizes=np.zeros(5)
#figure ,axes =plt.subplots(4,5,figsize=(18,12),dpi=200)
#ax = axes.flatten()
dic={}
temp={}
n=1
for i in range(4):
    s = 'level'+scities[i]
    for j in range(2016,2021):
        counts = [0]*5
        for k in range(5):
            if level[k] in df[s].loc[df['year'] == j].value_counts():
                counts[k] = df[s].loc[df['year'] == j].value_counts()[level[k]]
        
        
        #labels = level
        for k in range(5):
            sizes[k] = counts[k]*100/sum(counts)
        temp = {n:sizes}
        dic.update(temp)
        n+=1
        sizes=np.zeros(5)
dic
labels = ['Excellent', 'Good', 'Mild', 'Moderate', 'Severe']
figure ,axes =plt.subplots(4,5,figsize=(18,12),dpi=300)
ax = axes.flatten()
for i in range(1,21):
    labels = ['Excellent', 'Good', 'Mild', 'Moderate', 'Severe']
    colors = ['green', 'yellow', 'orange', 'red', 'purple']
    sizes = dic[i]
    labels=list(compress(labels, sizes>0))
    colors=list(compress(colors,sizes>0))
    sizes=sizes[sizes>0]
    ax[i-1].pie(sizes, labels=labels, autopct='%1.1f%%',colors=colors,
            shadow=True, startangle=90)
    ax[i-1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    ax[i-1].set_title('level'+scities[int((i-1)/5)] + '-' + str(df['time'][(i-1)%5*12])[0:4])
    
```
